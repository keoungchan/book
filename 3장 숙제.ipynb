### 상관관계가 무엇인지에 대해 조사하고 종류가 있다면 어떤것들이있는지 정리를하시오

상관 분석(Correlation Analysis) : 연속 변수로 된 두 변수간의 선형관계를 분석하는 기법 한쪽이 증가하면 다른 한쪽도 증가 , 감소 하는지를 나타냄

상관계수(Correlation Coefficient ) : 두 변수 사이의 선형적인 관계를 나타내며 한 변수의 변화가 다른 변수의 변화에 비례적으로 연관되어 있는 경우 선형 관계가 있다고 말한다.

피어슨(Pearson) 상관계수 : 연속형 자료와 연속형 자료가 어떤 상관관계를 가지는지 의미하는 수치며  두 변수가 모두 정규성을 따른다는 가정이 필요하다. +1은 양의 선형관계 0은 상관관계없음 -1 은 음의 선형 상관관계를 의미한다.

스피어만(Spearman) 상관게수 : 두 변수가 정규성을 따르지 않는 경우에 쓰이며 값에 순위를 매겨 그 순위에 대해 상관계수 값을 구하기 때문에 비모수적 방법이라고 할 수 있다.순위를 이용하기 때문에 연속형 변수가 아닌 순서형 변수인 경우에도 상관계수를 구할 수 있다.


### 모델을 평가하는 방법은? 어떤것들이있는지 조사

Precision(정밀도) : 모델이 True라고 한 것 중에 실제 정답이 True인 것

Recall(재현율) : 실제 정답이 True인 것 중에 모델이 True라고 한 것 

F1 Score : Precision과 Recall의 조화평균(0~1 값을 가지며 높을수록 좋음)

Accuracy(정확도) : 모델의 정확도

Confusion Matrix(혼동행렬)



True Positive : 정답은 True이고, 분류 결과도 True인 것 (정답)

False Positive: 정답은 False이고, 분류 결과는 True인 것 (오답)

False Negative: 정답은 True이고, 분류 결과는 False인 것 (오답)

True Negative: 정답은 False이고, 분류 결과도 False인 것 (정답)



을 예로 들엇을때 대각선의 색이 진할수록 잘 분류된것임

결정계수 : 평균제곱오차를 사용해 학습한 모델 예측의 적합도를 나타내며 보통 0과 1사이의 값으로 낸다 .그러나 예측값과 실제 데이터 사이의 오차가 너무 크면 음수값으로 나타내기도 한다.

AUC-ROC Curve
모델 뷴류 성능에 대한 측정그래프로 ROC(Receiver Operating Characeristic) 모든 임계값에서 분류 모델의 성능을 보여주는 그래프이며 AUC(Area Under the Curve)는 Roc 의 아래영역이다.



### 데이터 스케일러조사

StandardScaler :  feature의 평균을 0, 분산을 1로 변경하여 데이터의 정규분포를 만들며 회귀보다 분류에 더 효과적이다.

MinMaxScaler : 모든 feature가 0과 1사이에 위치하게 만듭니다.

MaxAbsScaler : 각 특성을 절대값이 0과 1사이가 되도록 스케일링하게 되며 모든 값은 -1과 1사이로 표현된다. 데이터가 양수일 경우 MinMaxScaler와 같다.

RobustScaler : 특성들의 중앙값을 0, IQR을 1로 스케일링한다. StandardScaler와 비슷하지만 이상치의 영향을 최소화합니다.
* IQR : 중간에 50%의 데이터들이 흩어진 정도를 의미하며  Q3(3사분위수(100분율의 75%)) - Q1(1사분위수(100분율의 25%)) = IQR 로 구할 수있다.


### bias와 variance

Bias(편항) : 학습된 분류기와 실제 값 사이의 제곱 에러를 말하며 bias가 high하면 모델이 정상적으로 학습을 마친 후 정확도가 기대치보다 떨어지는 것을 의미하게되고 즉 UnderFitting이 되었다고 말할 수 있다.

Variance(분산) : 분산은 지나치게 복잡한 모델로 인한 error를 나타내며 Variance가 high하면 정상적으로 학습을 마친 후 새로운 데이터 셋에 대해서는 그 결과가 매우 다르게 나온다는 것을 의미하게되며 즉 OverFitting이 되었다고 말할수 있다.



왼쪽 그림은 High Bias Low Variance를 뜻하며 모델이 낮은 variance와 높은 bias를 가질때 모델은 매우 안정적인 결과를 내놓으며 데이터가 달라도 비슷한 구역으로 예측값을 매핑하게된다. 문제점으로는 틀린 곳을 매핑한다는 것이 문제이다.

오른쪽그림은 High Variance Low Bias를 뜻하며 모델이 높은 variance와 낮은 bias를 가질 때 일부는 정확하게 매핑되지만 많은 데이터가 정확하게 예측하지 못하게된다.



bias와 variance를 잘 조율해야 하기 위해서는 오류를 최소화 해야한다. 최소화 하기 위해서는 bias와 variance의 합이 최소가 되는 지점을 찾으면 된다.
